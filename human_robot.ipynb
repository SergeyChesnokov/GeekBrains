```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
```

```python
lst
```

```python
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
pd.get_dummies(data)
```

Приведем в one hot вид без использования get dummies

```python
data.loc[data['whoAmI'] == 'human', 'human'] = '1'
data.loc[data['whoAmI'] == 'robot', 'human'] = '0'
data.loc[data['whoAmI'] == 'robot', 'robot'] = '1'
data.loc[data['whoAmI'] == 'human', 'robot'] = '0'

df_one_hot = data[['human', 'robot']]

df_one_hot
```
